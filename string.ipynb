Q1. Write a Python program to check if a given string is a valid email 
address. 
Ex: example@domain.com 

```python
x="example@domain.com" 
y=x.find('@')
z=x.find('.')
if y>z:
    print("Invalid")
else:
    print("Valid")
```

Q5. Write a Python program to find all unique characters in a given 
string. 
Input: "programming" 
Output: ['p', 'o', 'r', 'g', 'a', 'm', 'i', 'n']

```python
string=input("Enter String")
# str=string.split()
result=[]
for str in string:
    if str not in result:
        result.append(str)
print(result)
```

Write a Python program to count the number of words in a given 
string that start with a specific letter which was taken as a input. 
Input: "This is a test string for testing", 't' 
Output: 3

```python
string=input("Enter String")
letter=input("Enter letter")
count=0
str=string.lower()
word=str.split()
for word in word:
    if word.startswith(letter):
        count=count+1
print(count)
```

Q7. Write a Python program to split a string into a list of words and 
then join them back into a single string with a hyphen - as a 
separator. 
Input: "Python is awesome" 
Output: "Python-is-awesome"

```python
string=input("Enter String")
split=string.split(" ")
word="-".join(split)
print(word)
```
