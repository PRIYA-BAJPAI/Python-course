```python
#Empty tuple
tpl=()
tpl1=tuple()
print(tpl,tpl1)
print(type(tpl))
```

```python
#tuple without paranthesis
tpl=1,2,3,4,"hello"
print(tpl)
```

```python
#deleting tuple
tpl=1,2,3,4,"hello"
del(tpl)
print(tpl) #tuple is deleted hence we cannot print tpl
```

```python
#no updation of tuple
tpl=1,2,3,4,"hello"
tpl[2]=7 #it is not possible
print(tpl)
```

```python
#to change values in tuple 
#convert it to list first then update again convert it to list
tpl=(2,3,4,5,"hello")
tpl=list(tpl)
tpl[3]=90
tpl=tuple(tpl)
print(tpl)
```

```python
#count method
tpl=(2,3,4,5,3,"hello")
print(tpl.count(3))
```

```python
#sorted method
tpl=(1,2,3,4,5,6,6,6,7,8,9,9,1,4,3,5)
print(sorted(tpl))
```

```python
# reverse sorted
tpl=(1,2,3,4,5,6,6,6,7,8,9,9,1,4,3,5)
print(sorted(tpl,reverse=True))
```

```python
#slicing
tpl=(1,2,3,4,5,6,6,6,7,8,9,9,1,4,3,5)
tpl[-3::-6]
```
