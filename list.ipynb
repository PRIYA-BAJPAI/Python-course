```python
#List Slicing
lst=[3,12,15,17,5,8,9,2,11,15]
x=lst[ :5:-3]
y=lst[3:8:2]
z=lst[ :5:3]
print(x)
print(y)
print(z)

```

```python
#input list +5 odd element and -5 even element
list=eval(input("Enter list : "))
list[0:10:1]
for i in range(len(list)):
    if list[i]%2 != 0:
        list[i]=list[i]+5
    else:
        list[i]=list[i]-5
print(list)
```

```python
#write a program which will take input from user and print largest element of list
list=eval(input("Enter a list : "))
print(list)
max=0
for i in range(len(list)):
    if list[i]>max:
        max=list[i];
print(max)
```

```python
#write a program which will take input from user and print largest element and 2nd largest element of list
list=eval(input("Enter a list : "))
print(list)
max1=0
max2=0
for i in range(len(list)):
    if list[i]>max1:
        max2=max1
        max1=list[i]
    if list[i]>max2 and list[i]!=max1:
        max2=list[i]
print("Largest Element : ",max1)
print("2nd Largest Element : ",max2)
```
