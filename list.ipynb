```python
#List Creation
lst=[1,2,3,4,"hello",[7,8,9],5]
for i in lst:
    print(i);
```

```python
#List Slicing
lst=[3,12,15,17,5,8,9,2,11,15]
x=lst[ :5:-3]
y=lst[3:8:2]
z=lst[ :5:3]
print(x)
print(y)
print(z)

```

```python
#input list +5 odd element and -5 even element
list=eval(input("Enter list : "))
list[0:10:1]
for i in range(len(list)):
    if list[i]%2 != 0:
        list[i]=list[i]+5
    else:
        list[i]=list[i]-5
print(list)
```

```python
#write a program which will take input from user and print largest element of list
list=eval(input("Enter a list : "))
print(list)
max=0
for i in range(len(list)):
    if list[i]>max:
        max=list[i];
print(max)
```

```python
#write a program which will take input from user and print largest element and 2nd largest element of list
list=eval(input("Enter a list : "))
print(list)
max1=0
max2=0
for i in range(len(list)):
    if list[i]>max1:
        max2=max1
        max1=list[i]
    if list[i]>max2 and list[i]!=max1:
        max2=list[i]
print("Largest Element : ",max1)
print("2nd Largest Element : ",max2)
```

```python
#functions related to list in python
#append function
lst=[3,5,7,18,20,11,26]
lst.append("hello")
lst.append([3,4,5,6,7,8])
print(lst)
print(lst[-1][-3]) #accessing 6
```

```python
#insert function
lst=[3,5,7,18,20,11,26]
lst.insert(3,52)
print(lst)
#if we use -ve indexing then then element will be added towards left of the index given
#if index where element is to be added is -3 then element will be added to -4 index
```

```python
# append
lst=[3,5,7,18,20,11,26]
lst.extend([3,4,12,18,19]) #add as a element
lst.append([4,5,6,7,8])  #add as a list
lst+=[1,2,3,4]  #same as extend
print(lst)
```

```python
#delete
lst=[3,5,7,18,20,26]
del lst[3]
print(lst)

x=lst.pop(1)
print(lst)
print(x)

lst.remove(26)
print(lst)
```
