```python
#leap year
year=int(input("Enter year:"))
if year%4==0 and year%100!=0 or year%400==0:
    print("Leap yr")
else:
    print("Not leap yr")
```

```python
#season
month=int(input("Enter month"))
if month>=3 and month<=5:
    print("Spring")
elif month>=6 and month<=8:
    print("Summer")
elif month>=9 and month<=11:
    print("Autumn")
else:
    print("Winter")
```
