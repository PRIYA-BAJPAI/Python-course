increment even values by 5

```python
dt={1:2,3:5,5:7,8:34}
for k,v in dt.items():
    if v%2==0:
        dt[k]+=5
print(dt)
```

```python
#delete
dt={2:3,4:6,10:[23,3],'hello':5}
del dt[2]
print(dt)
print(dt.pop(4))  #value of key=4 will be visible
print(dt.popitem())
print(dt)
dt.clear()
print(dt)
```

wap which will print sum of all the keys whos values are the factor of 12

```python
dt={1:12,5:4,6:24,9:6}
sum=0
for k in dt.keys():
    if dt[k]%12==0 :
        sum = sum+k
print(sum)
```
